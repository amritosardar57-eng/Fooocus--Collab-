# 🚀 Fooocus + Auto Reconnect (Android Safe)
यह notebook Google Colab पर **Fooocus** चलाने के लिए है और इसमें **Auto Reconnect / No Idle script** already लगा है ताकि 5 मिनट में disconnect न हो।
👉 इसे पहले **Copy to Drive** करना मत भूलना।

```python
# ================================
# 🛠️ Prevent Google Colab Disconnect
# ================================
from IPython.display import HTML, display

def prevent_disconnect():
    display(HTML('''
    <script>
    function ClickConnect(){
        console.log("Preventing Google Colab from disconnecting…");
        document.querySelector("colab-toolbar-button#connect").click();
    }
    setInterval(ClickConnect, 60000);
    </script>
    '''))

prevent_disconnect()
```
